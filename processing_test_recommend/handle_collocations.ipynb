```python
def sort_ngramm_by_first_letter(ngramm_lists):
    ngramm_alph_sort_dict = {}
    for ngramm_list in ngramm_lists:
        for ngramm in ngramm_list:
            
            try:
                first_letter = ngramm[0][0]
                if first_letter not in ngramm_alph_sort_dict:
                    ngramm_alph_sort_dict[first_letter] = []
                    ngramm_alph_sort_dict[first_letter].append(ngramm)
                else:
                    ngramm_alph_sort_dict[first_letter].append(ngramm)
            except:
                print("SORTING FAILED", ngramm)
    return ngramm_alph_sort_dict
sort_ngramm_by_first_letter(colloc_dict['bigramms'])
```
