```python
word_suff = {"свежесть":"ест","трусость":"ост","ничтожество":"еств","хвастовство":"ств",
             "сдельщина":"щин","туретчина":"чин","реализм":"изм","седина":"ин",
             "радушие":"уш","спокойствие":"ств","кислота":"от","хрипота":"от",
             "белизна":"изн","наказание":"ан","терпение":"ен","развитие":"ит",
             "плавка":"ав","голодовка":"ов","бомбежка":"ёж","стилизация":"заци",
             "конкуренция":"енци","транспозиция":"озици","массаж":"аж","платёж":"ёж"}

def get_suff_list(word, suf_base):
    print(word, suf_base)
    parsed_word =  morph.parse(word)[0]
    suff = set()
    for lex in parsed_word.lexeme:
        print (lex[0])
        try:
            print (lex[0][lex[0].index(suf_base):],'\n')
            suff_i = lex[0][lex[0].index(suf_base):]
            suff.add(suff_i)
        except:
            print("SUBSTRING NOT FOUND")
    print("==============================================")
    return suff                

abstr_suff = set()    
for key, value in word_suff.items():
    suff_i_set = get_suff_list(key, value)
    abstr_suff = abstr_suff.union(suff_i_set)
    print(abstr_suff)
```

```python
import csv
#my_dict = {'1': 'aaa', '2': 'bbb', '3': 'ccc'}
with open('abstract_suff.csv', 'w+') as f:
    for key in abstr_suff:
        f.write("%s\n"%(key))
```
