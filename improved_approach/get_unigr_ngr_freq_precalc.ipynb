```python
import json
```

```python
with open ("./colloc/music_unigr_freq.json", 'r', encoding = 'utf-8') as f:
    unigr_freq = json.load(f)
    
with open ("./colloc/music_smart_colloc_freq_test.json", 'r', encoding = 'utf-8') as f:
    colloc_freq = json.load(f)
```

```python
colloc_freq['2']['российский паспорт']
```

```python
freq_colloc_lit = []
with open("enze_big_musician_colloc_lit.txt", "r", encoding = "utf-8") as f:
    not_found = 0
    for line_raw in f.readlines():
        line = line_raw[:-1]
        #print(line)
        line_len = len(line.split())
        if line in unigr_freq:
            freq_el = line + " " + str(unigr_freq[line])
            freq_colloc_lit.append(freq_el)
        elif line in colloc_freq[str(line_len)]:
            #print(line, "found in colloc")
            freq_el = line + " " + str(colloc_freq[str(line_len)][line])
            freq_colloc_lit.append(freq_el)
        else:
            print(line, "not found")
            not_found += 1
    print("not found", not_found)
        
```
