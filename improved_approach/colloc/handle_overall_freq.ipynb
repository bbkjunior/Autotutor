```python
import pandas as pd
import json
```

```python
df = pd.read_csv("freq_dict.csv",sep = '\t')
df.head(100)
```

```python
dict_colloc = df.to_dict('split')
```

```python
freq_dict = {}
for word_info in dict_colloc['data']:
    freq_dict[word_info[1]] = word_info[3]
```

```python
with open ("lyashevskaya_freq_dict.json","w", encoding = "utf-8") as f:
    json.dump(freq_dict, f, indent = 4, ensure_ascii = False)
```
