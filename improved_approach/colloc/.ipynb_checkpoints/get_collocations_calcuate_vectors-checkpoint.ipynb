```python
clean_words = []
for i in tqdm(range(len(filtered_tri))):
    posed_ngramm = filtered_tri.iloc[i]['trigram']
    clean_ngramm = []
    for w in posed_ngramm:
        word = w.split("_")[0]
        clean_ngramm.append(word)
    clean_ngramm = tuple(clean_ngramm)
    clean_words.append(clean_ngramm)
```
